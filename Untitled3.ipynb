```python
x = [1,2,3, 45, -5]
y = [2,4,6, 34, 27]
### Resizing the Graph

plt.figure(figsize=(5,3), dpi=300)


plt.plot(x,y, label='2x', color='red', marker='.', markersize=10, markeredgecolor='blue')


plt.title('YEP', fontdict={'fontname': 'League Gothic', 'fontsize': 30})

plt.xlabel('X NUMs')
plt.ylabel('Y NUMs')

# The numbers you put in the ticks session bellow will be the only ones in the graph

plt.xticks([1,2,3,45])
plt.yticks([2,4,6,34])
### SAVE THE GRAPH

#plt.savefig('mygraph.png', dpi=300)

plt.show()
```
